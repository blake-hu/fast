# Figure Plotting

```python
import matplotlib.pyplot as plt
import numpy as np
import pathlib
from datetime import datetime
```

```python
# Generate some data
x = [590592, 1770240, 2949888, 1242203, 6379099, 8326235, 82760793]
y = [7.96, 12.09, 15.93, 1125.00, 1136.00, 1221.80, 1408.00]
colors = [10, 10, 10, 20, 20, 20, 300]

# Create a scatter plot
plt.scatter(x, y, c=colors, cmap='cividis', alpha=0.75)
plt.xscale('log')
plt.yscale('log')

# Set the title and labels
plt.title('Random Scatter Plot')
plt.xlabel('Trainable Parameters (log scale)')
plt.ylabel('Training Time (log scale)')

# Save the plot to an image file
plot_root = pathlib.Path(f"./plots/")
plot_root.mkdir(parents=True, exist_ok=True)

save_file_id = datetime.now().strftime("%Y%m%d_%H%M%S")
filepath = plot_root / f'plot_{save_file_id}.png'
plt.savefig(filepath)

print(f"Scatter plot saved as '{filepath}'")
```
