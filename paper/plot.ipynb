# Figure Plotting

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import pathlib
from datetime import datetime
```

```python
# Generate some data
fast_x = [590592, 1770240, 2949888]
peft_x = [1242203, 6379099, 8326235]
full_x = [82760793]
fast_y = [7.96, 12.09, 15.93]
peft_y = [1125.00, 1136.00, 1221.80]
full_y = [1408.00]
colors = ['r', 'r', 'r', 'b', 'b', 'b', 'g']
markers = ['o', 'o', 'o', '^', '^', '^', 's']
plt.rcParams['font.family'] = 'Times New Roman'

# Create a scatter plot
plt.scatter(fast_x, fast_y, marker='o', s=100, label='FAST')
plt.scatter(peft_x, peft_y, marker='^', s=100, label='PEFT')
plt.scatter(full_x, full_y, marker='s', s=100, label='Full Fine-tune')

plt.legend(fontsize=12)
plt.xscale('log')
plt.yscale('log')

# Set the title and labels
plt.xlabel('Trainable Parameter Count', fontsize=12)
plt.ylabel('Training Time (s)', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)

# Save the plot to an image file
plot_root = pathlib.Path(f"./plots/")
plot_root.mkdir(parents=True, exist_ok=True)

save_file_id = datetime.now().strftime("%Y%m%d_%H%M%S")
filepath = plot_root / f'plot_{save_file_id}.png'
figure(figsize=(8, 6), dpi=300)
plt.savefig(filepath)

print(f"Scatter plot saved as '{filepath}'")
```
